```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")


df = pd.read_csv("IPL.csv")
df
```

Total number of matches per season

```python
matches_per_season = df[['match_id', 'season']].drop_duplicates().groupby('season').count()
matches_per_season = matches_per_season.rename(columns={'match_id': 'match_count'}).sort_values('match_count')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=matches_per_season.index, y='match_count', data=matches_per_season, palette='viridis')
plt.title('Total Matches per Season', fontsize=16)
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Number of matches hosted at each venue

```python
venue_match_count = df[['match_id', 'venue']].drop_duplicates()
venue_count = venue_match_count['venue'].value_counts()
```

```python
plt.figure(figsize=(10, 12))
sns.barplot(x=venue_count.values, y=venue_count.index, palette='magma')
plt.title('Number of Matches Hosted by Venue', fontsize=16)
plt.xlabel('Match Count')
plt.ylabel('Venue')
plt.tight_layout()
plt.show()
```

 Win Outcomes by Result Type (normal, superover, D/L method)

```python
df['result_type'].value_counts(dropna= False)
```

```python
# Replace NaN with 'unknown' for visualization clarity
result_counts = df[['match_id', 'result_type']].drop_duplicates()
result_counts['result_type'] = result_counts['result_type'].fillna('Unknown')
result_summary = result_counts['result_type'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:5]
plt.pie(result_summary, labels=result_summary.index, colors=colors, autopct='%.1f%%', startangle=140)
plt.title('Match Outcomes by Result Type', fontsize=14)
plt.show()
```

Number of Matches Played in Each City

```python
city_match_counts = df[['match_id', 'city']].drop_duplicates()
city_counts = city_match_counts['city'].value_counts().sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 12))
sns.barplot(x=city_counts.values, y=city_counts.index)
plt.title('Number of Matches Hosted by City', fontsize=16)
plt.xlabel('Match Count')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
# Select unique match-team pairs for both batting and bowling sides
batting_teams = df[['match_id', 'batting_team']].drop_duplicates()
bowling_teams = df[['match_id', 'bowling_team']].drop_duplicates()

# Rename columns for consistency
batting_teams.columns = ['match_id', 'team']
bowling_teams.columns = ['match_id', 'team']

# Combine and drop duplicates
all_teams = pd.concat([batting_teams, bowling_teams]).drop_duplicates()

# Count matches played per team
matches_per_team = all_teams['team'].value_counts().sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x=matches_per_team.values, y=matches_per_team.index, palette='Set2')
plt.title('Total Matches Played per Team', fontsize=16)
plt.xlabel('Number of Matches')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

Top 10 Teams by Total Runs Scored

```python
team_runs= df.groupby('batting_team')['runs_total'].sum().sort_values(ascending= False)
top_10_runs = team_runs.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_runs.values, y=top_10_runs.index, palette='Blues_r')
plt.title('Top 10 Teams by Total Runs Scored', fontsize=16)
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

Top 10 Teams by Total Wickets Taken

```python
team_wickets= df.groupby('bowling_team')['bowler_wicket'].sum().sort_values(ascending= False)
top_10_wickets= team_wickets.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_wickets.values, y=top_10_wickets.index, palette='OrRd')
plt.title('Top 10 Teams by Total Wickets Taken', fontsize=16)
plt.xlabel('Total Wickets')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

Toss Win vs Match Win Percentage

```python
match_outcomes = df[['match_id', 'toss_winner', 'match_won_by']].drop_duplicates()
match_outcomes['toss_win_and_match_win'] = match_outcomes['toss_winner'] == match_outcomes['match_won_by']
toss_stats = match_outcomes.groupby('toss_winner')['toss_win_and_match_win'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=toss_stats.index, y=toss_stats.values, palette='coolwarm')
plt.title('Match Win % When Toss is Won (Per Team)', fontsize=16)
plt.xlabel('Team')
plt.ylabel('Win % After Toss Win')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Team Run Rate

```python
team_run_rate = df.groupby('batting_team').apply(lambda x: x['runs_total'].sum() / (x['over'].max() + 1))
team_run_rate = team_run_rate.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=team_run_rate.values, y=team_run_rate.index, palette='spring')
plt.title('Average Team Run Rate', fontsize=16)
plt.xlabel('Run Rate')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
over_runs = df.groupby('over')['runs_total'].mean()

plt.figure(figsize=(10, 5))
sns.lineplot(x=over_runs.index, y=over_runs.values, marker='o', linewidth=2)
plt.title('Average Runs Scored per Over', fontsize=14)
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.xticks(range(0, 21))
plt.tight_layout()
plt.show()
```

```python
df_innings = df[['match_id', 'innings', 'batting_team', 'team_runs']].drop_duplicates()

plt.figure(figsize=(10, 6))
sns.histplot(df_innings['team_runs'], bins=30, kde=True, color='teal')
plt.title('Distribution of Team Scores per Innings')
plt.xlabel('Team Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
dismissals = df['wicket_kind'].dropna().value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=dismissals.values, y=dismissals.index, palette='Set1')
plt.title('Dismissal Types Frequency', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Dismissal Type')
plt.tight_layout()
plt.show()
```

```python
top_pairs = df[df['wicket_kind'].notna()]
top_pairs = top_pairs.groupby(['bowler', 'player_out']).size().reset_index(name='count')
top_pairs = top_pairs[top_pairs['count'] >= 5]
top_pivot = top_pairs.pivot(index='bowler', columns='player_out', values='count').fillna(0)

plt.figure(figsize=(12, 10))
sns.heatmap(top_pivot, cmap='Reds')
plt.title('Bowler vs Batter Dismissals (>=5 times)', fontsize=14)
plt.tight_layout()
plt.show()
```
